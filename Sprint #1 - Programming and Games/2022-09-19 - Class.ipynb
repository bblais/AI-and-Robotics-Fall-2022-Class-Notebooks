```python
%matplotlib inline
from mplturtle import *
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)
        
def draw_row(N):
    for i in range(N):
        square(10)
        forward(10)
```

```python
reset()
draw_row(8)

penup()

forward(100)
right(45)
forward(20)
right(30)
forward(50)


pendown()


draw_row(7)
```

```python
reset()
for i in [8,7,6,5,4]:
    draw_row(i)
    forward(100)
    print(i)
```
